```python
from sklearn.tree import DecisionTreeRegressor

class XGBoost:
 def __init__(self, n_trees=100, learning_rate=0.5, max_depth=13, reg_alpha=0, reg_lambda=0):
    self.n_trees = n_trees
    self.learning_rate = learning_rate
    self.max_depth = max_depth
    self.reg_alpha = reg_alpha
    self.reg_lambda = reg_lambda
    self.trees = []
    self.f0 = None

 def fit(self, X, y):
    # Normalize the input features
    X = (X - X.mean(axis=0)) / X.std(axis=0)
    # Round the target labels to integer values
    y = np.round(y)
    # Convert y to integer type
    y = y.astype(int)
    # Convert y to one-hot encoding
    y_onehot = np.zeros((y.shape[0], np.max(y) + 1))
    y_onehot[np.arange(y.shape[0]), y] = 1
    # Initialize the model
    self.f0 = np.mean(y_onehot, axis=0)
    pred = np.full((X.shape[0], y_onehot.shape[1]), self.f0)
    for i in range(self.n_trees):
      # Compute the gradients of the loss function with respect to the predictions
      grad = y_onehot - pred
      trees = []
      for j in range(y_onehot.shape[1]):
          # Compute the gradients of the loss function with respect to the outputs of the previous trees for each class
          grad_prev = grad[:, j] - self.reg_alpha * np.sign(pred[:, j]) -self.reg_lambda * pred[:, j]
          # Fit a new tree to the gradients of the current class
          tree = DecisionTreeRegressor(max_depth=self.max_depth)
          tree.fit(X, grad_prev)

          # Add the new tree to the model for each class
          trees.append(tree)
          # Update the predictions for the current class
          pred[:, j] += self.learning_rate * tree.predict(X)
          self.trees.append(trees)

 def predict_proba(self, X):
    # Initialize the predictions
    pred = np.full((X.shape[0], self.f0.shape[0]), self.f0)

    # Compute the predictions by summing the outputs of each tree for each class
    for i in range(self.n_trees):
      for j in range(len(self.trees[i])):
        pred[:, j] += self.learning_rate * self.trees[i][j].predict(X)

    # Apply the softmax function to get the class probabilities
    pred = np.exp(pred - np.max(pred, axis=1, keepdims=True))
    return pred / np.sum(pred, axis=1, keepdims=True)

 def predict(self, X):
    # Predict the class with the highest probability
    return np.argmax(self.predict_proba(X), axis=1)

# Menghitung akurasi model
akurasi = accuracy_score(y_test, y_pred)
print("Akurasi:", akurasi)

# Menghitung confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)

# Menampilkan confusion matrix
print("Confusion Matrix:")
print(conf_mat)

# Menampilkan classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import numpy as np

class AdaBoost:
    def __init__(self, n_clf=50):
        self.n_clf = n_clf

    def fit(self, X, y):
        m, n = X.shape
        W = np.full(m, 1/m)
        self.clfs = []
        self.alphas = []

        for _ in range(self.n_clf):
            clf, error, y_pred = self._build_stump(X, y, W)
            alpha = 0.5 * np.log((1 - error) / (error + 1e-10))
            W = W * np.exp(-alpha * y * y_pred)
            W = W / np.sum(W)

            self.clfs.append(clf)
            self.alphas.append(alpha)

    def _build_stump(self, X, y, W):
        m, n = X.shape
        best_clf = {}
        best_error = float('inf')
        best_y_pred = np.zeros(m)

        for feature_idx in range(n):
            for threshold in np.unique(X[:, feature_idx]):
                for direction in ['less_than', 'greater_than']:
                    y_pred = self._stump_predict(X, feature_idx, threshold, direction)
                    error = np.sum(W[y_pred != y])

                    if error < best_error:
                        best_clf['feature_idx'] = feature_idx
                        best_clf['threshold'] = threshold
                        best_clf['direction'] = direction
                        best_error = error
                        best_y_pred = y_pred

        return best_clf, best_error, best_y_pred

    def _stump_predict(self, X, feature_idx, threshold, direction):
        y_pred = np.ones(X.shape[0])
        if direction == 'less_than':
            y_pred[X[:, feature_idx] < threshold] = -1
        else:
            y_pred[X[:, feature_idx] > threshold] = -1
        return y_pred

    def predict(self, X):
        m = X.shape[0]
        final_output = np.zeros(m)

        for alpha, clf in zip(self.alphas, self.clfs):
            final_output += alpha * self._stump_predict(X, clf['feature_idx'], clf['threshold'], clf['direction'])

        return np.sign(final_output)

# Menghitung akurasi model
akurasi = accuracy_score(y_test, y_pred)
print("Akurasi:", akurasi)

# Menghitung confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)

# Menampilkan confusion matrix
print("Confusion Matrix:")
print(conf_mat)

# Menampilkan classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor

class GradientBoostingClassifier:
    def __init__(self, n_trees=100, learning_rate=0.5, max_depth=13, reg_alpha=0, reg_lambda=0):
        self.n_trees = n_trees
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.reg_alpha = reg_alpha
        self.reg_lambda = reg_lambda
        self.trees = []
        self.f0 = None

    def fit(self, X, y):
        X = (X - X.mean(axis=0)) / X.std(axis=0)
        y = np.round(y).astype(int)
        y_onehot = np.zeros((y.size, y.max() + 1))
        y_onehot[np.arange(y.size), y] = 1

        self.f0 = y_onehot.mean(axis=0)
        pred = np.full((X.shape[0], y_onehot.shape[1]), self.f0)
        for i in range(self.n_trees):
            trees = []
            grad = y_onehot - pred
            for j in range(y_onehot.shape[1]):
                grad_prev = grad[:, j] - self.reg_alpha * np.sign(pred[:, j]) - self.reg_lambda * pred[:, j]
                tree = DecisionTreeRegressor(max_depth=self.max_depth)
                tree.fit(X, grad_prev)
                trees.append(tree)
                pred[:, j] += self.learning_rate * tree.predict(X)
            self.trees.append(trees)

    def predict_proba(self, X):
        pred = np.full((X.shape[0], self.f0.shape[0]), self.f0)
        for trees in self.trees:
            for j, tree in enumerate(trees):
                pred[:, j] += self.learning_rate * tree.predict(X)
        pred = np.exp(pred - np.max(pred, axis=1, keepdims=True))
        return pred / np.sum(pred, axis=1, keepdims=True)

    def predict(self, X):
        return np.argmax(self.predict_proba(X), axis=1)

# Menghitung akurasi model
akurasi = accuracy_score(y_test, y_pred)
print("Akurasi:", akurasi)

# Menghitung confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)

# Menampilkan confusion matrix
print("Confusion Matrix:")
print(conf_mat)

# Menampilkan classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

# Data
metrics = ['Presisi', 'Recall', 'F1-Score', 'Akurasi']
xgboost = [86.7, 76.2, 86.7, 86.9]
adaboost = [79.8, 59.3, 79.9, 81.2]
gradient_boosting = [86.7, 74.4, 86.6, 87.0]
soft_voting = [91.3, 82.7, 91.4, 91.6]

# Transposed data for the new layout
data = [
    xgboost,
    adaboost,
    gradient_boosting,
    soft_voting
]

# Algorithms for the x-axis
algorithms = ['XGBoost', 'AdaBoost', 'Gradient Boosting', 'Soft Voting']

# Setting the positions of the bars on the x-axis
bar_width = 0.2
indices = range(len(algorithms))

# Colors for the metrics
metric_colors = ['#ff007f', '#40e0d0', '#ff00ff', '#dc143c']  # Corresponding to Presisi, Recall, F1-Score, Akurasi

# Plotting
plt.figure(figsize=(10, 6))

# Plotting the bars with labels for each metric
for i, metric in enumerate(metrics):
    metric_data = [algorithm[i] for algorithm in data]
    bar_positions = [x + (bar_width * i) - (bar_width * 1.5) for x in indices]
    bars = plt.bar(bar_positions, metric_data, width=bar_width, color=metric_colors[i], label=metric)

    # Adding value labels on top of each bar
    for bar, value in zip(bars, metric_data):
        plt.text(bar.get_x() + bar.get_width() / 2, value + 1, str(value), ha='center', color='black', fontsize=8)

# Setting the labels for the x-axis
plt.xticks(ticks=[i - bar_width / 2 for i in indices], labels=algorithms)

# Adding labels and title
plt.ylabel('Persentase (%)')
plt.title('Perbandingan Kinerja Model')

# Moving the legend to the right side of the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()
```
