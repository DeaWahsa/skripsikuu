```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score

class LightGBM:
    def __init__(self, learning_rate=0.1, n_estimators=100, max_depth=3):
        self.learning_rate = learning_rate
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.trees = []

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def gradient(self, y_true, y_pred):
        return y_true - self.sigmoid(y_pred)

    def fit(self, X, y):
        # Inisialisasi prediksi awal
        pred = np.zeros(X.shape[0])

        for _ in range(self.n_estimators):
            # Hitung gradien
            grad = self.gradient(y, pred)

            # Fit pohon ke gradien
            tree = DecisionTreeRegressor(max_depth=self.max_depth)
            tree.fit(X, grad)

            # Update prediksi dengan mempertimbangkan learning rate
            pred += self.learning_rate * tree.predict(X)

            # Tambahkan pohon ke dalam daftar pohon
            self.trees.append(tree)

    def predict_proba(self, X):
        # Inisialisasi prediksi awal
        pred = np.zeros(X.shape[0])

        # Hitung prediksi dengan menjumlahkan output setiap pohon
        for tree in self.trees:
            pred += self.learning_rate * tree.predict(X)

        # Gunakan sigmoid untuk mendapatkan probabilitas kelas 1
        proba = self.sigmoid(pred)
        return np.column_stack((1 - proba, proba))

    def predict(self, X):
        # Prediksi kelas berdasarkan probabilitas
        return np.argmax(self.predict_proba(X), axis=1)

# Inisialisasi model LightGBM
lgbm = LightGBM(learning_rate=0.1, n_estimators=100, max_depth=3)

# Latih model pada data latih
lgbm.fit(X_train, y_train)

# Lakukan prediksi pada data uji
y_pred = lgbm.predict(X_test)

# Hitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi:", accuracy)
```

```python
import numpy as np

def softmax(z):
    # Menghitung softmax
    exp_scores = np.exp(z)
    return exp_scores / np.sum(exp_scores, axis=1, keepdims=True)

def compute_gradient(y_pred, y_true):
    # Menghitung gradien loss menggunakan cross-entropy loss
    n_samples = y_pred.shape[0]
    grad = softmax(y_pred)
    grad[range(n_samples), y_true] -= 1
    grad /= n_samples
    return grad
```

```python
class DecisionNode:
    def __init__(self, feature_idx=None, threshold=None, value=None, left=None, right=None):
        self.feature_idx = feature_idx
        self.threshold = threshold
        self.value = value
        self.left = left
        self.right = right

class DecisionTree:
    def __init__(self, max_depth=1):
        self.max_depth = max_depth

    def compute_gini(self, y):
        # Menghitung gini impurity
        classes = np.unique(y)
        n_samples = len(y)
        gini = 1.0
        for cls in classes:
            p = np.sum(y == cls) / n_samples
            gini -= p ** 2
        return gini

    def split(self, X, y, feature_idx, threshold):
        # Memisahkan dataset berdasarkan feature dan threshold
        left_mask = X[:, feature_idx] < threshold
        right_mask = ~left_mask
        left_X, left_y = X[left_mask], y[left_mask]
        right_X, right_y = X[right_mask], y[right_mask]
        return left_X, left_y, right_X, right_y

    def find_best_split(self, X, y):
        # Mencari pemisahan terbaik berdasarkan gini impurity
        n_features = X.shape[1]
        best_gini = float('inf')
        best_feature_idx, best_threshold = None, None
        for feature_idx in range(n_features):
            thresholds = np.unique(X[:, feature_idx])
            for threshold in thresholds:
                left_X, left_y, right_X, right_y = self.split(X, y, feature_idx, threshold)
                gini = (len(left_y) * self.compute_gini(left_y) + len(right_y) * self.compute_gini(right_y)) / len(y)
                if gini < best_gini:
                    best_gini = gini
                    best_feature_idx = feature_idx
                    best_threshold = threshold
        return best_feature_idx, best_threshold

    def build_tree(self, X, y, depth=0):
        # Rekursif membangun decision tree
        if depth >= self.max_depth or len(np.unique(y)) == 1:
            leaf_value = np.mean(y, axis=0)
            return DecisionNode(value=leaf_value)

        feature_idx, threshold = self.find_best_split(X, y)
        left_X, left_y, right_X, right_y = self.split(X, y, feature_idx, threshold)
        left = self.build_tree(left_X, left_y, depth + 1)
        right = self.build_tree(right_X, right_y, depth + 1)
        return DecisionNode(feature_idx=feature_idx, threshold=threshold, left=left, right=right)

    def predict(self, x, node):
        # Memprediksi kelas menggunakan decision tree
        if node.value is not None:
            return node.value
        if x[node.feature_idx] < node.threshold:
            return self.predict(x, node.left)
        else:
            return self.predict(x, node.right)
```

```python
class XGBoostClassifier:
    def __init__(self, n_estimators=100, learning_rate=0.1, max_depth=1):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.estimators = []

    def fit(self, X, y):
        # Fitting XGBoost dengan menggunakan weak learner berbasis decision tree
        y_pred = np.zeros_like(y, dtype=float)

        for _ in range(self.n_estimators):
            gradient = compute_gradient(y_pred, y)
            tree = DecisionTree(max_depth=self.max_depth)
            tree_root = tree.build_tree(X, gradient)
            self.estimators.append(tree_root)

            update = self.learning_rate * tree.predict(X, tree_root)
            y_pred += update

    def predict_proba(self, X):
        # Memprediksi probabilitas kelas menggunakan ensemble model
        y_pred = np.zeros((X.shape[0], len(self.estimators)))

        for i, estimator in enumerate(self.estimators):
            y_pred[:, i] = self.learning_rate * estimator.predict(X, estimator)

        return softmax(y_pred)

    def predict(self, X):
        # Memprediksi kelas menggunakan ensemble model
        proba_xgb = self.predict_proba(X)
        return np.argmax(proba_xgb, axis=1)
```

```python
class LightGBMClassifier:
    def __init__(self, n_estimators=100, learning_rate=0.1, max_depth=1):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.estimators = []

    def fit(self, X, y):
        # Fitting LightGBM dengan menggunakan weak learner berbasis decision tree
        y_pred = np.zeros_like(y, dtype=float)

        for _ in range(self.n_estimators):
            gradient = compute_gradient(y_pred, y)
            tree = DecisionTree(max_depth=self.max_depth)
            tree_root = tree.build_tree(X, gradient)
            self.estimators.append(tree_root)

            update = self.learning_rate * tree.predict(X, tree_root)
            y_pred += update

        self.y_pred = y_pred

    def predict_proba(self, X):
        # Memprediksi probabilitas kelas menggunakan ensemble model
        y_pred = np.zeros((X.shape[0], len(self.estimators)))

        for i, estimator in enumerate(self.estimators):
            y_pred[:, i] = estimator.predict(X, estimator)

        return softmax(self.learning_rate * y_pred)

    def predict(self, X):
        # Memprediksi kelas menggunakan ensemble model
        proba_lgb = self.predict_proba(X)
        return np.argmax(proba_lgb, axis=1)
```

```python
import numpy as np
import xgboost as xgb
import lightgbm as lgb
from sklearn.ensemble import GradientBoostingClassifier

# Latih klasifier XGBoost
xgb_clf = xgb.XGBClassifier()
xgb_clf.fit(X_train, y_train)

# Latih klasifier LightGBM
lgb_clf = lgb.LGBMClassifier()
lgb_clf.fit(X_train, y_train)

# Latih klasifier Gradient Boosting
gb_clf = GradientBoostingClassifier()
gb_clf.fit(X_train, y_train)

# Step 2: Tentukan kelas Firefly Optimization

class FireflyOptimizer(object):
    """
    Catatan
    ----------------
    Posisi update dari firefly mengikuti persamaan di bawah ini:
        xi(t+1) = xi(t) + beta*exp(-gamma*dist**2)*(xj(t) - xi(t)) + alpha(t)*epsilon(t)
    """

    def __init__(self, limits, alpha=1.0, beta=1.0, gamma=0.01, maxIter=100):
        # Parameter model
        self.alpha = alpha
        self.beta = beta
        self.gamma = gamma

        # Hyper-parameter
        self.limits = limits
        self.maxIter = maxIter

    def optimize(
        self, fireflies, measure="Euclid", reduction=0.97, tol=1.0e-1
    ):
        """Menjalankan Firefly optimizer"""
        # Dapatkan fitur input
        num_firefly = len(fireflies)
        center = np.mean(fireflies, axis=0)

        # Pilih metode pengukuran jarak
        if measure == "Euclid":
            self.metric = self.calc_euclidean_distance
        elif measure == "Manhattan":
            self.metric = self.calc_manhattan_distance
        else:
            print("METRIC JARAK SALAH")
            raise ValueError

        # Hitung daya tarik (attractiveness)
        attractiveness = np.array(
            [-self.metric(firefly, center) for firefly in fireflies]
        )

        # Optimasi
        curr_fireflies = fireflies.copy()
        next_fireflies = np.zeros_like(fireflies)
        for n in range(self.maxIter):
            for i, j in np.ndindex((num_firefly, num_firefly)):
                if i == j:  # Tidak ada daya tarik pada diri sendiri
                    continue
                elif attractiveness[i] < attractiveness[j]:
                    next_fireflies[i] = self.move(
                        curr_fireflies[i], curr_fireflies[j]
                    )

            # Periksa konvergensi
            mse = np.mean(
                np.sqrt(np.sum((curr_fireflies - next_fireflies) ** 2.0))
            )
            if mse < tol:
                break

            # Hitung daya tarik (attractiveness)
            curr_fireflies = next_fireflies
            attractiveness = np.array(
                [-self.metric(firefly, center) for firefly in curr_fireflies]
            )

            # Lakukan relaksasi dengan menerapkan perpindahan infinitesimal
            curr_fireflies = self.relax(curr_fireflies, delta=0.1)  # Kurangi randomness
            self.alpha *= reduction

        return curr_fireflies

    def move(self, xi, xj):
        """Memindahkan firefly ke arah yang menarik"""
        # Perpindahan
        dr = self.metric(xi, xj)

        # Daya tarik (attraction)
        attraction = self.beta * np.exp(-self.gamma * dr**2.0)

        # Skala langkah perpindahan
        scaling = (
            (self.limits[1] - self.limits[0])
            * self.alpha
            * (np.random.random(xi.shape) - 0.5)  # Menggunakan shape dari xi
        )

        # Perbarui posisi
        xi += np.clip(
            attraction * (xj - xi) + scaling, self.limits[0], self.limits[1]
        )

        return xi

    def relax(self, fireflies, delta=0.01):
        """Menyebar fireflies secara acak"""
        return np.array(
            [
                np.random.uniform(firefly - delta, firefly + delta)
                for firefly in fireflies
            ]
        )

    def calc_euclidean_distance(self, a, b):
        """Menghitung jarak Euclidean"""
        return np.sqrt(np.sum((a - b) ** 2))

    def calc_manhattan_distance(self, a, b):
        """Menghitung jarak Manhattan"""
        return np.sum(np.abs(a - b))

# Step 3: Implementasikan Soft Voting dengan Firefly Optimization

# Gabungkan klasifier dan bobot terkait
def soft_voting_firefly_predict(X, classifiers, weights):
    votes = np.zeros((X.shape[0], len(classifiers)))

    for i, clf in enumerate(classifiers):
        votes[:, i] = clf.predict_proba(X)[:, 1]  # Gunakan probabilitas kelas positif saja

    weighted_votes = np.dot(votes, weights)
    return weighted_votes

# Tentukan klasifier dan bobot terkait
classifiers = [xgb_clf, lgb_clf, gb_clf]

# Buat instance FireflyOptimizer dan jalankan optimisasi
limits = (0, 1)  # Batas bobot, asumsikan mereka berada antara 0 dan 1
firefly_opt = FireflyOptimizer(limits)
optimized_weights = firefly_opt.optimize(np.random.rand(len(classifiers)), measure="Euclid", reduction=0.97, tol=1.0e-1)

# Normalisasi bobot agar jumlahnya menjadi 1
optimized_weights /= np.sum(optimized_weights)

# Step 4: Evaluasi Soft Voting Classifier dengan bobot teroptimasi

# Tentukan klasifier dan bobot terkait
classifiers = [xgb_clf, lgb_clf, gb_clf]

# Hitung akurasi masing-masing klasifier
accuracies = [accuracy_score(y_test, clf.predict(X_test)) for clf in classifiers]

# Invers akurasi untuk mendapatkan bobot awal
initial_weights = 1 / np.array(accuracies)

# Buat instance FireflyOptimizer dan jalankan optimisasi
limits = (0, 1)  # Batas bobot, asumsikan mereka berada antara 0 dan 1
maxIter = 500  # Tingkatkan jumlah iterasi untuk lebih banyak kesempatan optimisasi
firefly_opt = FireflyOptimizer(limits, maxIter=maxIter)
optimized_weights = firefly_opt.optimize(initial_weights, measure="Euclid", reduction=0.97, tol=1.0e-1)

# Normalisasi bobot agar jumlahnya menjadi 1
optimized_weights /= np.sum(optimized_weights)

# Step 5: Evaluasi akurasi pada data uji

# Gunakan bobot teroptimasi untuk membuat prediksi
y_pred_prob = soft_voting_firefly_predict(X_test, classifiers, optimized_weights)

# Konversi probabilitas menjadi prediksi kelas dengan threshold 0.5
y_pred = (y_pred_prob >= 0.5).astype(int)

# Hitung akurasi
accuracy = np.mean(y_pred == y_test)

# Cetak bobot teroptimasi dan akurasi
print("Bobot terbaik hasil Algoritma Firefly:", optimized_weights)
print("Akurasi Soft Voting dengan bobot terbaik:", accuracy)
```
